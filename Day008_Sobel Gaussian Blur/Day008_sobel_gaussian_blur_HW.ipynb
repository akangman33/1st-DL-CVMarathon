實作模糊與邊緣檢測

- 透過 Gaussian Filter 實作模糊操作
- 透過 Sobel Filter 實作邊緣檢測

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 8)

img = cv2.imread('../data/lena.png')
```

## 邊緣檢測

### 比較 Sobel 如果在 uint8 的情況下做會 overflow 的狀況

```python
# 轉為灰階圖片
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 對 x 方向以包含負數的資料格式 (cv2.CV_16S) 進行 Sobel 邊緣檢測
img_sobel_x = cv2.Sobel(img_grey, cv2.CV_16S, 1, 0, ksize=3)

# 對 x 方向依照比例縮放到所有數值都是非負整數
img_sobel_x = cv2.convertScaleAbs(img_sobel_x)

# 對 x 方向直接以非負整數的資料格式 (uint8) 進行 Sobel 邊緣檢測
img_sobel_x_uint8 = cv2.Sobel(img_grey, cv2.CV_8U, 1, 0, ksize=3)

#　組合 + 顯示圖片
plt.subplot(1,3,1)
plt.axis('off')
plt.title('Original')
plt.imshow(img_grey, cmap='gray')

plt.subplot(1,3,2)
plt.axis('off')
plt.title('16S Sobel')
plt.imshow(img_sobel_x, cmap='gray')

plt.subplot(1,3,3)
plt.axis('off')
plt.title('8U Sobel')
plt.imshow(img_sobel_x_uint8, cmap='gray')
# img_show = np.hstack((img_grey, img_sobel_x, img_sobel_x_uint8))
# while True:
#     # 比較 Sobel 邊緣檢測的過程中針對負數操作的不同產生的差異
#     cv2.imshow('Edge Detection', img_show)
#     k = cv2.waitKey(0)
#     if k == 27:
#         cv2.destroyAllWindows()
#         break
```

### 比較一次與兩次計算偏微分的結果

```python
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 求一次導數取得邊緣檢測結果
img_sobel_x = cv2.Sobel(img_grey, cv2.CV_16S, 1, 0)

# 求二次導數取得邊緣檢測結果
img_sobel_xx = cv2.Sobel(img_grey, cv2.CV_16S, 2, 0)

img_sobel_x = cv2.convertScaleAbs(img_sobel_x)
img_sobel_xx = cv2.convertScaleAbs(img_sobel_xx)

#　組合 + 顯示圖片
plt.subplot(1,3,1)
plt.axis('off')
plt.title('Original')
plt.imshow(img_grey, cmap='gray')

plt.subplot(1,3,2)
plt.axis('off')
plt.title('1D Sobel')
plt.imshow(img_sobel_x, cmap='gray')

plt.subplot(1,3,3)
plt.axis('off')
plt.title('2D Sobel')
plt.imshow(img_sobel_xx, cmap='gray')

# img_show = np.hstack((img_grey, img_sobel_x, img_sobel_xx))
# while True:
#     cv2.imshow('Edge Detection', img_show)
#     k = cv2.waitKey(0)
#     if k == 27:
#         cv2.destroyAllWindows()
#         break
```
