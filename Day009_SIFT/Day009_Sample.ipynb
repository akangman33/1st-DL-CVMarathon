# 範例

取得 SIFT 特徵

- 轉成灰階圖片
- 需要額外安裝 OpenCV 相關套件

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParamsDefault['figure.figsize'] = (10, 8)

img = cv2.imread('../data/lena.png')
```

```python
# 轉灰階圖片
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 建立 SIFT 物件
SIFT_detector = cv2.xfeatures2d.SIFT_create()

# 取得 SIFT 關鍵點位置
keypoints_gray = SIFT_detector.detect(img_gray, None)

#　畫圖 + 顯示圖片
img_show_gray = cv2.drawKeypoints(img_gray, keypoints_gray, img)

plt.axis('off')
plt.imshow(img_show_gray)
# while True:
#     cv2.imshow('SIFT', img_show)
#     k = cv2.waitKey(0)
#     if k == 27:
#         cv2.destroyAllWindows()
#         break
```

```python
img = cv2.imread('../data/lena.png')
img_r = img[...,-1]
img_g = img[...,-2]
img_b = img[...,-3]


# 建立 SIFT 物件
SIFT_detector = cv2.xfeatures2d.SIFT_create()

# 取得 SIFT 關鍵點位置
keypoints = SIFT_detector.detect(img_r, None)

#　畫圖 + 顯示圖片
img_show = cv2.drawKeypoints(img_r, keypoints, img)
plt.subplot(1,4,1)
plt.axis('off')
plt.imshow(img_show)

img = cv2.imread('../data/lena.png')
SIFT_detector = cv2.xfeatures2d.SIFT_create()
keypoints = SIFT_detector.detect(img_g, None)
img_show = cv2.drawKeypoints(img_g, keypoints, img)
plt.subplot(1,4,2)
plt.axis('off')
plt.imshow(img_show)

img = cv2.imread('../data/lena.png')
SIFT_detector = cv2.xfeatures2d.SIFT_create()
keypoints = SIFT_detector.detect(img_b, None)
img_show = cv2.drawKeypoints(img_b, keypoints, img)
plt.subplot(1,4,3)
plt.axis('off')
plt.imshow(img_show)


img = cv2.imread('../data/lena.png')
plt.subplot(1,4,4)
plt.axis('off')
plt.imshow(img_show_gray)
plt.tight_layout()
# while True:
#     cv2.imshow('SIFT', img_show)
#     k = cv2.waitKey(0)
#     if k == 27:
#         cv2.destroyAllWindows()
#         break
```

```python
img = cv2.imread('../data/lena.png')
img_r = img[...,-1]
img_g = img[...,-2]
img_b = img[...,-3]

# 建立 SIFT 物件
SIFT_detector = cv2.xfeatures2d.SIFT_create()

# 取得 SIFT 關鍵點位置
keypoints = SIFT_detector.detect(img_r, None)
print("Red Feature Points:", len(keypoints))


SIFT_detector = cv2.xfeatures2d.SIFT_create()
g_keypoints = SIFT_detector.detect(img_g, None)
keypoints.extend(g_keypoints)
print("Green Feature Points:", len(g_keypoints))


SIFT_detector = cv2.xfeatures2d.SIFT_create()
b_keypoints = SIFT_detector.detect(img_b, None)
keypoints.extend(b_keypoints)
print("Blue Feature Points:", len(SIFT_detector.detect(img_b, None)))

img_show = cv2.drawKeypoints(img, keypoints, img)


plt.axis('off')
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))

plt.figure()
plt.axis('off')
plt.imshow(img_show_gray)
# while True:
#     cv2.imshow('SIFT', img_show)
#     k = cv2.waitKey(0)
#     if k == 27:
#         cv2.destroyAllWindows()
#         break
```
